```python
import matplotlib.pyplot as pylt

pylt.plot([0,2,3,4],[0,8,27,64],label="y=x^2",color="black",marker='o')
pylt.plot([4.4,1/2,1/3,1/4],[0.202,2,3,4],label="y = 1 / x", color="red", marker='x')

pylt.grid(True)
pylt.xlabel("x-axis")
pylt.ylabel("y-axix")
pylt.title("functions")
pylt.legend()
pylt.savefig("function")
pylt.show()
```

```python
import matplotlib.pyplot as pylt
import numpy as np

#example of vector space
#field itself a vector space
transfor = np.array([[0,1],[1,0]])
swapnum = np.array([2,3])
swap = np.dot(transfor,swapnum)

pylt.figure(figsize=(2,2))
pylt.plot(swap, marker = 'o')
pylt.show()





#take Zp(where p is prime number)
#R ** n , n may be natural natural

```
